```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_dict = {'Restrictions':[2,2,3,3,4,4,5,5],
'Type':['SeqVec','Random','SeqVec','Random','SeqVec','Random','SeqVec','Random'],
'AUC':[0.87,0.78, 0.85, 0.75, 0.71, 0.74, 0.62, 0.65],
'Test size':[11609, 11609, 14467, 14467, 2793, 2793, 439, 439]}
```

```python
df = pd.DataFrame(result_dict)

pd.pivot_table(df, index = 'Restrictions', columns = 'Type', values 
= 'AUC').plot(kind = 'bar')
plt.title('AUC of shortest path restricted test set')
plt.xlabel('Shortest path Restriction')
plt.ylabel('AUC')
```
